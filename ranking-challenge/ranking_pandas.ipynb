```python
# import the dependencies
import pandas as pd
import numpy as np
```

```python
# create a list of strings and numbers as ranks, shuffle them 
a = np.arange(10)
np.random.shuffle(a)
b = ['Hitman', 'God of War', 'Prince of Persia', 'Shadow of Colossus']
c = [np.random.choice(b) for x in range(len(a))]
```

```python
# create a dataframe, assign index a name, move it to a column
rank_df = pd.DataFrame(a, c, columns={"Rank"})
rank_df.index.names = ["Name"]
rank_df.reset_index(inplace=True)
rank_df
```

```python
# group by Name to find the highest rank of each game 
high_ranks =  rank_df.groupby("Name", as_index=False).min().rename(columns={"Rank":"Highest Rank"})
high_ranks.head()
```

```python
# finally create a dataframe that has highest rank of each game assigned to it as a new column 
final_ranks = pd.merge(rank_df, high_ranks, on="Name", how="left")
final_ranks 
```
