```python
# import the dependencies
import pandas as pd
```

```python
# import the excel and display
rank_df = pd.read_excel("RankData.xlsx")
rank_df
```

```python
# group by Name to find the highest rank of each game 
high_ranks =  rank_df.groupby("Name", as_index=False).min().rename(columns={"Rank":"Highest Rank"})
high_ranks.head()
```

```python
# merge on Name column to get the desired results. 
final_ranks = pd.merge(rank_df, high_ranks, on="Name", how="left")
final_ranks 
```
