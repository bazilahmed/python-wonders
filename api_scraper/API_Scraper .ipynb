```python
# Program to import jobs data
# Import the dependencies 
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send the request and parse the html response 
# url = "https://www.linkedin.com/jobs/search/?keywords=data&location=United%20States&locationId=us%3A0"
# url = "https://www.linkedin.com/jobs/search?keywords=Data+Analyst&locationId=us:0&start=0&count=25&applyLogin=false&trk=jobs_jserp_pagination_1"
url = "https://www.indeed.com/jobs?q=data+analyst"

```

```python
# open the url and create an request object
header = {"User Agent" : "Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"}
    
r = requests.get(url, header)
r.status_code

soup = BeautifulSoup(r.content, "html.parser")

print(soup.prettify())
```

```python
job_results = soup.find_all('div', class_='jobsearch-SerpJobCard row result')
len(job_results)
```

```python
# for row in job_results:
#     print(row.prettify())
    
print(job_results[4].prettify())
```

```python
# retrieve the job title
titles = []
for row in job_results:
    try:
        job_title = row.find('a', class_="jobtitle turnstileLink").text.strip()
        titles.append(job_title)
    except AttributeError:    
        job_title = row.find('a', class_="turnstileLink").text.strip()
        titles.append(job_title)
titles
```

```python
# retrieve the job summary 
summaries = []
for row in job_results:
    summary = row.find('div', class_="paddedSummary").text.strip()
    summaries.append(summary)

summaries
```

```python
# retrieve the company
companies = []
for row in job_results:
    company = row.find("span", class_="company").text.strip()
    companies.append(company)

companies
```

```python
# retrive salary if provided 
salaries = []
for row in job_results:
    try:
        salary = row.find("span", class_="salary no-wrap").text.strip()
        salaries.append(salary)
    except AttributeError:
        salaries.append("Not available")

salaries
```

```python
# retrieve the location
locations = []
for row in job_results:
    try:
        location = row.find("div", class_="location").text.strip()
        locations.append(location)
    except AttributeError:
        location = row.find("span", class_="location").text.strip()
        locations.append(location)
        
locations
```

```python
# create a dataframe and output to csv
job_results_df = pd.DataFrame({ 'Title': titles,
                              'Summary': summaries,
                              'Company': companies,
                              'Location': locations,
                              'Salary': salaries})

job_results_df = job_results_df[['Title', 'Summary', 'Company', 'Location', 'Salary']]

job_results_df
```
